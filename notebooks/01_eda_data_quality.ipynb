This notebook profiles the Kaggle shipment-pricing dataset and explains data quality findings after every chart.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.preprocessing import load_base_file, preprocess_dataset

raw_df = load_base_file()
df = preprocess_dataset(raw_df)
df.shape, df.isna().mean().sort_values(ascending=False).head(15)
```

## Interpretation

```python
vendor_volume = df.groupby('vendor', as_index=False)['line_item_quantity'].sum().sort_values('line_item_quantity', ascending=False).head(10)
sns.barplot(data=vendor_volume, x='line_item_quantity', y='vendor')
plt.title('Top Vendors by Shipment Volume')
plt.show()
```
